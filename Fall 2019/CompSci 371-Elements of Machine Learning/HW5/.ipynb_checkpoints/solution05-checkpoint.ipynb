## Part 1: Numerical Differentiation

```python
import numpy as np


def function(x):
    return np.sin(x)


def derivative(x):
    return np.cos(x)


def forward_difference(f, x, h=1.e-5):
    assert h > 0, 'h must be positive'
    return (f(x + h) - f(x)) / h


def error(method, f, d, x):
    h_values = np.logspace(-16, -1, num=101)
    return [np.abs(method(f, x, h) - d(x)) for h in h_values], h_values, method.__name__
```

### Problem 1.1

### Solution

```python
import matplotlib.pyplot as plt
%matplotlib inline

x_errors = [(error(forward_difference, function, derivative, x), x) for x in (0.0, 1.0)]
plt.figure(figsize=(8, 8))
for e in x_errors:
    plt.loglog(e[0][1], e[0][0], label='x={}'.format(e[1]))
plt.legend()
plt.xlabel('h')
plt.ylabel('total approximation error')
plt.show()
```

### Problem 1.2 (Exam-Style)

### Solution

**Low $h$:** The plot for $x = 0$ has zeros everywhere for $h < 10^{-8}$ or so. In this range, the dominant error is the rounding error. However, both $f(x + h)$ and $f(x)$ are close to zero when $x=0$ and $h$ is small. More specifically, $f(0)$ is exactly zero, and $f(h)$ is nearly equal to $h$ when $h$ is small. Therefore the difference is well-conditioned (the two numbers being subtracted are no greater than their difference), leading to nearly no rounding error.

**High $h$:** The plot for $x = 0$ is between 1 and 6 orders of magnitude smaller than that for $x = 1$ for values of $h$ greater than $10^{-8}$ or so. In this range, the dominant error is the truncation error. However, the second derivative of the sine function at $x = 0$ is zero. Since the coefficient $M$ of $h$ in the first-order term of the truncation error for the forward difference method depends on the second derivative of $f$, the error term $Mh$ is zero for $x = 0$. Therefore, the first-order term of truncation error vanishes, leading to a higher-order (and therefore much smaller) truncation error for $x=0$.

_A quantitative note_ (not needed for full credit): In the high-$h$ range, the plot for $x=0$ swings from about $10^{-16}$ to about $10^{-3}$, about 13 orders of magnitude. That for $x=1$ swings from about $10^{-8}$ to about $10^{1.5}$, about 6.5 orders of magnitude. Thus, the blue plot has twice the slope of the orange plot on a logarithmic scale. Since the growth of the truncation error is linear for $x = 1$ (from the discussion in the introduction, with $M > 0$), this implies that the truncation error for $x=0$ grows quadratically. This conclusion would be borne out by theoretical analysis.

### Problem 1.3 (Exam-Style)

### Solution

The main downward slope in the left part of the plot is due to rounding error when taking the difference $f(x+h) - f(x)$. The oscillations around this slope, on the other hand, are caused by the much smaller rounding errors in the values of $f(x+h)$ and $f(x)$ themselves. These values are computed with finite arithmetic, and are therefore approximate.

The truncation error is algebraic rather than numerical, and is approximately a polynomial, because of Taylor's theorem. This is why the plots are smooth for larger $h$.

### Problem 1.4

### Solution

```python
def central_difference(f, x, h=1.e-5):
    assert h > 0, 'h must be positive'
    return (f(x + h) - f(x - h)) / h / 2.0


def four_point_difference(f, x, h=1.e-5):
    assert h > 0, 'h must be positive'
    return (-f(x + 2 * h) + 8 * f(x + h) - 8 * f(x - h) + f(x - 2 * h)) / h / 12.0


x_0 = 1.0
errors = [error(formula, function, derivative, x_0)
          for formula in (forward_difference, central_difference, four_point_difference)]

plt.figure(figsize=(8, 8))
for e in errors:
    plt.loglog(e[1], e[0], label=e[2])
plt.legend()
plt.xlabel('h')
plt.ylabel('total approximation error')
plt.show()
```

### Problem 1.6

### Solution

The following function computes and prints approximate slopes for the part of the plots dominated by the truncation error:

```python
def slope(method, f, d, x, hs):
    def log_value(z, h):
        return np.log10(np.abs(method(f, z, h) - d(z)))

    step, rise = np.log10(hs[1]) - np.log10(hs[0]), log_value(x, hs[1]) - log_value(x, hs[0])
    return rise / step


slopes = [slope(formula, function, derivative, 1.0, (1.e-2, 1.e-1))
          for formula in (forward_difference, central_difference, four_point_difference)]

for s in slopes:
    print('{:.3f}'.format(s))
```

The slopes are approximately in the ratios 1, 2, 4. Therefore, since slope in a log-log plot is proportional to degree, the truncation error for central difference is order $O(h^2)$ and that for the four-point difference is $O(h^4)$. These results would be borne out by theoretical analysis.

### Problem 2.2 (Exam-Style)

### Solution

From the class notes, the point $\mathbf{p}$ is

$$
\mathbf{p} = \beta \mathbf{n} \;\;\;\text{where}\;\;\;
\beta = - \frac{b}{\|\mathbf{w}\|} \;\;\;\text{and}\;\;\;
\mathbf{n} = \frac{\mathbf{w}}{\|\mathbf{w}\|}
$$

so that

$$
\mathbf{p} = - b \frac{\mathbf{w}}{\|\mathbf{w}\|^2} \;.
$$

The point $\mathbf{n}$ defined above has unit norm and is perpendicular to $\ell$.

In the real plane, if $\mathbf{n} = [n_1, n_2]^T$, then the two unit-norm vectors perpendicular to $\mathbf{n}$, and therefore parallel to $\ell$, are $\mathbf{v}_{1,2} = \pm [n_2, -n_1]^T$, so that

$$
\mathbf{v}_{1,2} = \pm \frac{1}{\|\mathbf{w}\|} \left[\begin{array}{r}
w_2\\- w_1
\end{array}\right]
$$

where $\mathbf{w} = [w_1, w_2]^T$.

### Problem 2.3 (Exam-Style)

### Solution

Since the given parametric representation is canonical, the point $\mathbf{p}$ is the point on $\ell$ that closest to the origin, and the vector $\mathbf{p}$ is therefore perpendicular to $\ell$.

Since $\mathbf{w}$ is a unit-norm vector perpendicular to $\ell$, it is one of

$$
\pm \frac{\mathbf{p}}{\|\mathbf{p}\|}\;.
$$

The coefficient $b$ is the negative of the distance of $\ell$ from the origin, and therefore

$$
b = -\|\mathbf{p}\|\;.
$$

All we have left to do is to determine which of the two possible signs for $\mathbf{w}$ is consistent with this choice for $b$. The point $\mathbf{p}$ (obtained from the parametric equation when $\alpha = 0$) must satisfy the equation for $\ell$, that is,

$$
b + \mathbf{w}^T \mathbf{p} = 0 \;\;\;\text{or}\;\;\;
\mathbf{w}^T \mathbf{p} = \|\mathbf{p}\|\;,
$$

from which we see that the plus sign must be used from the two choices for $\mathbf{w}$. In conclusion,

$$
\mathbf{w} = \frac{\mathbf{p}}{\|\mathbf{p}\|} \;\;\;\text{and}\;\;\;
b = -\|\mathbf{p}\|\;.
$$

## Part 3: The Logistic-Regression Loss

### Problem 3.1

### Solution

```python
def regression_loss(y, alpha):
    score = 1 / (1 + np.exp(-alpha))
    return - y * np.log(score) - (1 - y) * np.log(1 - score)


alpha_range, samples = (-3, 3), 101
alphas = np.linspace(alpha_range[0], alpha_range[1], samples)
plt.figure(figsize=(8, 6))
for label in (0, 1):
    r = regression_loss(label, alphas)
    plt.plot(alphas, r, label='label {}'.format(label))
plt.plot((0, 0), plt.ylim(), 'k', lw=0.5)
plt.legend()
```

### Problem 3.2 (Exam-Style)

### Solution

We have

$$
\rho(\alpha) = - \log p = \log (1 + e^{-\alpha}) \;.
$$

This function is differentiable everywhere and has first and second derivatives

$$
\rho'(\alpha) = -\frac{e^{-\alpha}}{1 + e^{-\alpha}} = - \frac{1}{e^{\alpha} + 1}
$$

and

$$
\rho''(\alpha) = \frac{e^{\alpha}}{(e^{\alpha} + 1)^2} \;.
$$

Both the numerator and the denominator of the fraction above are positive, and therefore

$$
\rho''(\alpha) > 0 \;\;\;\text{for all}\ \alpha\in \mathbb{R}\;.
$$

Therefore, the function $\rho(\alpha)$ is convex.

### Problem 3.3 (Exam-Style)

### Solution

From the previous problem, we see that

$$
\lim_{\alpha\rightarrow -\infty}\rho'(\alpha) = -1 \;.
$$

In other words, the slope of $\rho$ tends to $-1$ as $\alpha$ tends to $-\infty$, and the rate of growth of $\rho$ is therefore linear in $\alpha$. The approximating polynomial would have degree 1.
